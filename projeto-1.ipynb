```python
import numpy as np
import matplotlib.pyplot as plt

#formula da função
def f(X):
    return np.sin(X)

#definindo intervalo
Xi = -np.pi
Xf = np.pi

#criação de um array para a função
X = np.linspace(Xi, Xf) 

#eixo Y vai ser a função em relação a X
Y = f(X)

#criação do gráfico
plt.plot(X, Y, label='f(X) = sen(X) ')
plt.title('f(X)')

plt.axhline(y = 0, color='black') 
plt.axvline(x = 0, color='black')

plt.xlabel('X')
plt.ylabel('Y')

#pintando a área a ser calculada
plt.fill_between(X, Y, color='blue', alpha=0.2)

#apresentação do gráfico na tela
plt.grid()
plt.legend()
plt.show()
```

```python
#área vai ser o resultado do Somátorio dos retangulos
area = 0
DX = 0.0000001

#criando varias linha hiper finas para encontrar a área da função
for X in np.arange(Xi, Xf, DX):
    
        #somando todas as linhas criadas para achar o numero mais proximo da área exata
    area += f(X)*DX

print(f'Area_retangulos ≈ {area}')
```

```python
#area vai ser o resultado do somatorio de todos os trapezios
area = 0
DX = 0.0001

#criação dos trapézios
for X in np.arange(Xi, Xf, DX):
    area += f(X) + f(X+DX)

#dividindo pela metade o resultado da area
area = area * DX * 0.5
print(f'Area_Trapezio ≅ {area}')
```

```python
#ponto medio do inicio e fim do intervalo
DX = (Xf + Xi) * 0.5

#formula do método
Area = f(Xi) + 4*f(DX) + f(Xf)
Area = DX/3 * Area

print(f'Area_Simpson = {Area}')
```
